```python
#Import dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

#File to Load
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

#Read the Mouse and Drug Data and the Clinical Trial Data
trial_data = pd.read_csv(mouse_drug_data_to_load)
mouse_data = pd.read_csv(clinical_trial_data_to_load)


#Combine the data into a single dataset
combined_df = trial_data.merge(mouse_data, on = 'Mouse ID')

combined_df = combined_df.sort_values(["Timepoint", "Tumor Volume (mm3)"])
combined_df = combined_df.reset_index()
combined_df.drop(combined_df.columns[0], axis=1, inplace=True)

#Display the data table for preview
combined_df.head()

```

## Tumor Response to Treatment

### Mean Tumor Volume Data

```python
#Store Mean Tumor Volume Data Grouped by Drug and Timepoint, Consolidate / Clean
mtv_data = pd.DataFrame(combined_df.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)'])
mtv_data = mtv_data.unstack(level = 0)
mtv_data.columns = mtv_data.columns.get_level_values(1)

#Preview dataframe: Mean Tumor Volume of Tumor by Drug and Timepoint
mtv_data
```

### Tumor Volume Data (Standard Error)

```python
#Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
volume_sem = pd.DataFrame(combined_df.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)'])

#Convert to DataFrame
volume_sem = volume_sem.unstack(level=0)
volume_sem.columns = volume_sem.columns.get_level_values(1)

#Preview DataFrame

volume_sem
```

```python
#plot
plt.figure(figsize = (15,10))
plt.title('Tumor Volume Over Time', fontdict = {'fontsize': 18, 'fontweight': 'bold'})
plt.xlabel('Time (Days)', fontdict = {'fontsize': 15})
plt.ylabel('Tumor Volume (mm3)', fontdict = {'fontsize': 15})
plt.xticks(np.arange(0, mtv_data.index.max()+3 , 5)) 
plt.grid()

#location of separators for timepoint
plt.xlim(0, mtv_data.index.max() + 1) 

#markers
markers = ['o', 's', '^', 'd','>','h','+','x','*','.']

xvals = mtv_data.index

count = 0

#plot each tumor volume column with standard error
for c in mtv_data.columns:
    plt.errorbar(xvals, 
                 mtv_data[c], 
                 volume_sem[c],
                 linestyle = '--', 
                 marker =  markers[count], 
                 capthick = 1, 
                 capsize = 3)
    count += 1

#legend
lg = plt.legend(numpoints = 2,
                frameon = True, 
                markerscale = 1.5, 
                edgecolor = 'black', 
                fontsize = '17', 
                framealpha = 1)

#Export and preview figure
plt.savefig("Tumor Response to Treatment.png")
plt.show()
```

## Metastatic Response to Treatment

### Mean Metastatic Site Data

```python
#Store the Mean Met. Site Data Grouped by Drug and Timepoint 
#Group by treatment for metastatic site counts
#Convert to DataFrame
#Consolidate / Celan
#Preview
mean_spread = pd.DataFrame(combined_df.groupby(['Drug', 'Timepoint']).mean()['Metastatic Sites'])

mean_spread = mean_spread.unstack(level = 0)
mean_spread.columns = mean_spread.columns.get_level_values(level = 1)
mean_spread
```

### Metastatic Site Data (Standard Error)

```python
spread_err = pd.DataFrame(combined_df.groupby(['Drug', 'Timepoint']).sem()['Metastatic Sites'])
spread_err = spread_err.unstack(level = 0)
spread_err.columns = spread_err.columns.get_level_values(level = 1)
spread_err
```

```python
#Plot
plt.figure(figsize = (15,10))
plt.title('Metastasization During Treatment', fontdict = {'fontsize': 18, 'fontweight': 'bold'})
plt.xlabel('Length of Treatmnent (Days)', fontdict = {'fontsize': 15})
plt.ylabel('Number of Metastatics Sites', fontdict = {'fontsize': 15})
plt.xticks(np.arange(0,mean_spread.index.max() + 3 ,5))
plt.xlim(0, mean_spread.index.max()+1)
plt.ylim(0, mean_spread.max().max() + spread_err.max().max() + .1)
plt.grid()

#creates plot for each treatment, using different markers for each

count = 0 
xvals = mean_spread.index #timepoints

for c in mean_spread:
    plt.errorbar(xvals, 
                 mean_spread[c], 
                 spread_err[c], 
                 linestyle = '--', 
                 marker = markers[count],
                 capthick = 1, 
                 capsize = 3)
    count += 1
    
#legend
lg = plt.legend(numpoints = 2, 
                frameon = True, 
                markerscale = 1.5, 
                edgecolor = 'black', 
                fontsize = '17', 
                framealpha = 1)
#Save Figure / Preview
plt.savefig("Metastasization During Treatment.png")
plt.show()
```

## Survival Rates

```python
#Store count of mice grouped by 'drug' and 'timepoint' in dataframe
#Preview dataframe
mice_count = combined_df.groupby(['Drug', 'Timepoint']).count()['Mouse ID']
mice_count = mice_count.unstack(level = 0)
mice_count
```

```python
#plot
plt.figure(figsize = (15,10))
plt.title('Survival Rate During Treatment', fontdict = {'fontsize': 18, 'fontweight': 'bold'})
plt.xlabel('Time (Days)', fontdict = {'fontsize': 15})
plt.ylabel('Survival Rate (%)', fontdict = {'fontsize': 15})
plt.xlim(0, mice_count.index.max())
plt.grid()

xvals = mice_count.index

count = 0 

for c in mice_count:
    yvals = mice_count[c]/mice_count.loc[0,c] * 100 
    plt.plot(xvals, 
             yvals, 
             linestyle = '--', 
             marker =  markers[count],
             )
    count += 1
    
#legend
lg = plt.legend(numpoints = 2, 
                frameon = True, 
                markerscale = 1.5, 
                edgecolor = 'black', 
                fontsize = '17', 
                framealpha = 1)
#Export / Preview Figure
plt.savefig("Survival Rate During Treatment.png")
plt.show()
```

## Summary

```python
#Calculate the percent changes for each drug
summary_data = (mtv_data.loc[45, :] - mtv_data.loc[0, :])/mtv_data.loc[0, :] * 100

#Display the data to confirm
summary_data
```

### Summary bar graph

```python
#Store all Relevant Percent Changes into a Tuple
fig, ax = plt.subplots(figsize=(15, 5))
summary_pass = dict((k, v) for k, v in summary_data.items() if v <= 0)
summary_pass = sorted(summary_pass.items())
x_pass, y_pass = zip(*summary_pass)
rects1 = ax.bar(x_pass, y_pass, color="g", align="center")

summary_fail = dict((k, v) for k, v in summary_data.items() if v >= 0)
summary_fail = sorted(summary_fail.items())
x_fail, y_fail = zip(*summary_fail)
rects2 = ax.bar(x_fail, y_fail, color="r", align="center")

#Orient widths. Add labels, tick marks, etc. 
plt.title('Tumor Volume Dynamics over Treatment', fontdict = {'fontsize': 12, 'fontweight': 'bold'})
plt.ylabel('Tumor Volume Change (%)')
plt.axhline(y=0, color = 'black') #adds a horizontal line at zero
xlabels = summary_data.index
ax.grid()
count = 0




#Labeling
rects = ax.patches
y_values = y_pass + y_fail
for rect, y_ in zip(rects, y_values):
    height = rect.get_height()
    if height > 0:
        ax.text(rect.get_x() + rect.get_width() / 2, 0+1, str(int(y_))+"%", ha='center', va='bottom',color='w')
    else:
        ax.text(rect.get_x() + rect.get_width() / 2, 0-5, str(int(y_))+"%", ha='center', va='bottom',color='w')

#Save figure
plt.savefig("Tumor Dynamics.png")

#Show figure
fig.show()
```

## Comments

Among the 10 drugs tested in this clinical trial, Capomulin and Ramicane showed the greatest reduction in both tumor volume and metastasization. Moreover, Capomulin and Ceftamin had the most positive mice outcome results. The other drugs (Ceftamin, Infubinol, Ketapril, Naftisol, Propriva, Stelasyn, Zoniferol) had continued tumor growth and higher mortality rates. Infubinol, Propiva, Zoniferol, and Ceftamin showed lower tumor growth than the placebo but also showed sustained metastasization and poor survival outcomes. Due to their promising outcomes, Capomulin and Ramicane can be recommended for further testing.
